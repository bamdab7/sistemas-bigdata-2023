```python
# Assign columns to new variables (DataFrames)
supervivintes = df[['Survived','Sex']]
#supervivintes.head()
```

```python
# Build Dataframes with many Series
nomes = df.Name
idades = df.Age
sexo = df.Sex

novo_df = pd.concat([nomes,idades,sexo],axis=1)
novo_df.head()

# S a modo de exemplo, pois sera moito mis fcil...
# novo_df = df['Name','Age','Sex']
```

```python
# Write a CSV file
novo_df.to_csv('output/titanic_redux.csv')
```

```python
# Read the new file
novo_csv = pd.read_csv('output/titanic_redux.csv')
novo_csv.head(3)
#novo_csv.sample(5)
```

```python
# The indexes are duplicated <- the save operation includes the default indexes as data
# We can use the first column as index while loading
novo_csv = pd.read_csv('output/titanic_redux.csv',index_col=0)
novo_csv.head()
```

```python
# Options in CSV writing
# Different separators (default = ,)
novo_df.to_csv('output/titanic_redux.csv', sep=';')
```
